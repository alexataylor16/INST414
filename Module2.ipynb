```python
import pandas as pd
from scipy.spatial.distance import cdist
```

```python
df = pd.read_csv('/content/1976-2020-president.csv')
df
```

```python
year_df = df[df['year'] == 2020]
year_df
```

```python
final_df = year_df[['state', 'candidate', 'candidatevotes', 'totalvotes']]
final_df
```

```python
final_df.value_counts('candidate')
```

```python
pivot_df = final_df.pivot_table(index='state', columns='candidate', values='candidatevotes', aggfunc='sum')
pivot_df = pivot_df.fillna(0)
pivot_df = pivot_df.drop('BLANK VOTES', axis=1)
pivot_df = pivot_df.drop('UNDERVOTES', axis=1)
pivot_df = pivot_df.drop('OVERVOTES', axis=1)
pivot_df = pivot_df.drop('VOID', axis=1)
pivot_df
```

```python
df_norm = pivot_df.divide(pivot_df.sum(axis=1), axis=0)
df_norm
```

```python
df_norm.sum(axis=1)
```

```python
third_df = pivot_df
third_df['total'] = third_df.sum(axis=1)
third_df['primary'] = third_df['TRUMP, DONALD J.'] + third_df['BIDEN, JOSEPH R. JR']
third_df['third'] = third_df['total'] - third_df['primary']
third_df['third_proportion'] = third_df['third'] / third_df['total']
top_states = third_df[['third_proportion']].sort_values(by='third_proportion', ascending=False)
top_states.head(5)
```

```python
states = ['UTAH', 'ALASKA', 'COLORADO', 'IDAHO', 'WYOMING']
for s in states:
  state = df_norm.loc[s]
  cosine_distances = cdist(df_norm, [state], metric='cosine')
  query_distances = list(zip(df_norm.index, cosine_distances))
  print(s)
  for similar_state, similarity_score in sorted(query_distances, key=lambda x: x[1], reverse=False)[:11]:
    print(f"{similar_state}: {similarity_score}")
```
